```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine.data = wine.iloc[:,1:]
wine.data
```

```python
# converting into array 
Wine=wine.data.values
Wine
```

```python
# Normalizing the Numerical data 
Wine_norm=scale(Wine)
```

```python
Wine_norm
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(Wine_norm)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
plt.plot(range(1,14), pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
pca_values
```

```python
pcdata=pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])], axis = 1)
```

```python
pcdata
```

# Kmean clustering

```python

from sklearn.cluster import KMeans
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalizing the data got from first three pca values
df_norm = norm_func(pcdata)


df_norm.head(10)  # Top 10 rows
```

```python
model=KMeans(n_clusters=3) 
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row 
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object 
wine['clust']=md # creating a  new column and assigning it to new column 
wine.head()
```
